# 4. Finding the outliers

### Making a box plof of a few columns to visualize the outliers

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(df.experimental_proprty)

plt.subplot(1, 3, 2)
sns.boxplot(df.nHetero)

plt.subplot(1, 3, 3)
sns.boxplot(df.nX)

plt.show()
```

### Removing the outliers using the Interquartile Range

```python
for col in df.columns:
    
    Q1 = np.percentile(df[col], 25)
    Q3 = np.percentile(df[col], 75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    median = df[col].median()  #Replacing the outliers with the median for each column

    df[col] = np.where((df[col] <= lower_bound) | (df[col] >= upper_bound), median, df[col])
```

```python
df.shape
#We can see that the shape of the data frame is preserved
```

# 5. Finding the correlated variables

```python
correlation_matrix = df.corr()
threshold = 0.994  #This threshold is set by trial and error
highly_correlated_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            col1 = correlation_matrix.columns[i]
            col2 = correlation_matrix.columns[j]
            highly_correlated_pairs.append((col1, col2, correlation_matrix.iloc[i, j]))

print("\nHighly correlated pairs:")
for pair in highly_correlated_pairs:
    print(f"{pair[0]} and {pair[1]}: {pair[2]:.2f}") 
```

### Plotting a heatmap of the highly correlated variables

```python
sns.heatmap(correlation_matrix)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# 6. Finding the highly correlated variables with the target

```python
target = 'experimental_proprty'
corr_matrix = df.corr()
target_correlation = corr_matrix[target]

threshold = 0.315  #This threshold is set by trial and error
significant_correlations = target_correlation[abs(target_correlation) > threshold]

print("\nFeatures with significant correlation with experimental_proprty:")
print(significant_correlations)
```

```python
significant_features = significant_correlations.index
sig_corr_matrix = corr_matrix.loc[significant_features, significant_features]
```

### Plotting a heat map of the highly correlated features with the target variable

```python
sns.heatmap(sig_corr_matrix)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# 7. 2D PCA

### Standardizing the data

```python
from sklearn.preprocessing import StandardScaler

significant_correlations = significant_correlations.index.tolist()

X = df[significant_correlations]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Computing the PCA Mathematically

```python
cov_matrix = np.cov(X_scaled.T)  #Computing the Covariance Matrix

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)  #Computing the Eigenvalues and Eigenvectors

sorted_indices = np.argsort(eigenvalues)[::-1]  #Sorting the Eigenvalues and Eigenvectors
eigenvalues_sorted = eigenvalues[sorted_indices]
eigenvectors_sorted = eigenvectors[:, sorted_indices]

top_2_eigenvectors = eigenvectors_sorted[:, :2]  #Selecting the Top 2 Principal Components

X_pca = X_scaled.dot(top_2_eigenvectors) #Plotting the PCA
```

```python
# Plotting the PCA result

plt.figure(figsize=(5, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df[target]) 
plt.colorbar(label=target)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Plot')
plt.show()
```


### The PCA plot shows a spread of data points with some central clustering. No clear distinct clusters are present.


```python
#Performing PCA using the Python in-built package
from sklearn.decomposition import PCA

pca = PCA(n_components=2) 
X_pca = pca.fit_transform(X_scaled)
```

```python
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

# 2. Finding any surprising relations in the data

## Performing K-Means clustering on the PCA transformed data

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_pca)

plt.figure(figsize=(5, 4))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters)
plt.colorbar(label='Cluster Label')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA with k-Means Clustering')
plt.show()
```
